```python
import pandas as pd
import re
import pickle
from typing import Dict, List, TypeVar
from nltk.tokenize import word_tokenize
from nltk import tokenize as tknz
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('train_tweets.csv')
data.head(2)
```

#### Предпроцессинг текста

```python
def lower_txt(input_txt:str)->str:
    out_txt = input_txt.lower()
    return out_txt
```

```python
def sub_txt(pattern:str, input_txt:str)->str:
    out_txt = re.sub(pattern, ' ', input_txt)
    return out_txt
```

```python
def replace_txt(replace_dict:Dict[str, str], input_txt:str)->str:
    list_txt = []
    for word in input_txt.split():
        if word in replace_dict.keys():
            word = replace_dict[word]
        list_txt.append(word)
    out_txt = ' '.join(list_txt)
    return out_txt
```

```python
def word_len(input_txt:str)->str:
    return ' '.join([w for w in input_txt.split() if len(w)>1])
```

```python
def preprocessing_txt(input_txt:str)->str:
    x = sub_txt(r'@[\w]*', input_txt)    # удалим @user из всех твитов
    x = lower_txt(x)                     # изменим регистр твитов
    x = replace_txt(apostrophe_dict, x)  # заменим сокращения с апострофами
    x = replace_txt(short_word_dict, x)  # заменим сокращения на их полные формы
    x = replace_txt(emoticon_dict, x)    # заменим эмотиконы
    x = sub_txt(r'[^\w\s]', x)           # заменим пунктуацию на пробелы
    x = sub_txt(r'[^a-zA-Z0-9]', x)      # заменим спец. символы на пробелы
    x = sub_txt(r'[^a-zA-Z]', x)         # заменим числа на пробелы
    x = word_len(x)                      # ограничим длинну слов
    return x
```

```python
data['tweet_pre'] = data['tweet'].apply(lambda x: preprocessing_txt(x))
data.head(2)
```

#### Токенизация и стоп слова

```python
data['tweet_token'] = data['tweet_pre'].apply(lambda x: tknz.word_tokenize(x)) 
```

```python
def stop_words(input_txt:List[str])->List[str]:
    stop_words = stopwords.words('english')
    list_txt = [word for word in input_txt if not word in stop_words]
    return list_txt
```

```python
data['tweet_token_filtered'] = data['tweet_token'].apply(lambda x: stop_words(x))  
```

#### Стемминг

```python
Stem = TypeVar("PorterStemmer")
def stemmer_(stemmer:Stem, words:List[str])->List[str]:
    list_txt = [stemmer.stem(word) for word in words]    
    return list_txt
```

```python
stemmer = PorterStemmer()
data['tweet_stemmed'] = data['tweet_token_filtered'].apply(lambda x: stemmer_(stemmer, x)) 
```

#### Лемматизация

```python
Lemm = TypeVar("WordNetLemmatizer")
def lemmatizer_(lemmatizer:Lemm, words:List[str])->List[str]:
    list_txt = [lemmatizer.lemmatize(word) for word in words] 
    return list_txt
```

```python
lemmatizer = WordNetLemmatizer()
data['tweet_lemmatized'] = data['tweet_token_filtered'].apply(lambda x: lemmatizer_(lemmatizer, x))
```
